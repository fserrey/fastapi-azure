```python
import pandas
from summarize import summarize
import warnings
warnings.filterwarnings("ignore")
import nltk
nltk.download(['stopwords', 'punkt'])
```

```python
_path = "/mnt/c/Users/Francisco Serrano/Downloads/Textos.xlsx"

import pandas as pd

textos = pd.read_excel(_path)
```

```python
display(textos.head())
```

```python
df = textos

df['word_count'] = df['Descripción'].apply(lambda x: len(str(x).split(' ')))
df['char_count'] = df['Descripción'].apply(lambda x: len(x))
display(df.head())
```

```python
df_ = df[df.char_count > 325]
```

```python
df_['resumen'] = df_['Descripción'].apply(lambda x: summarize(str(x),sentence_count=2, language='spanish'))
df_['resumen_wordCount'] = df_['resumen'].apply(lambda x: len(str(x).split(' ')))
df_['resumen_charNum'] = df_['resumen'].apply(lambda x: len(x))
```

```python
display(df_)
```

```python
df_.to_excel(_path+"output.xlsx")
```
